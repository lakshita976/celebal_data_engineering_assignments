## Creating Quantity Column

```python
df["quantity"] = np.random.randint(1, 6, size=len(df))
```

## Cleaning Price Column

```python
df["final_price"] = (
    df["final_price"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False)
    .astype(float)
)
```

```python
df["final_price"].dtype
```

## Creating Derived Column

```python
df["total_amount"] = df["final_price"] * df["quantity"]
```
