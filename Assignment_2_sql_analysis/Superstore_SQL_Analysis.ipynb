## Setup - Import Libraries (SQLite)

```python
import sqlite3
import pandas as pd
```

---
## Step 1 - Load Dataset into SQL Database
> This step reads the CSV file and creates a SQL table called `db`.

```python
CSV_PATH = 'Sample_Superstore_Underscore.csv'

df = pd.read_csv(CSV_PATH)

conn = sqlite3.connect('superstore.db')

df.to_sql('db', conn, if_exists='replace', index=False)

print("Data loaded into SQLite table")
```

---
## Step 2 - Explore the Table

### 2a. Total Rows in Table
> Counts the total number of records imported.

```python
pd.read_sql_query("""
    SELECT COUNT(*) AS total_rows
    FROM db
""", conn)
```

### 2b. Sample Data (LIMIT 10)
> Displays the first 10 records for quick inspection.

```python
pd.read_sql_query("""
    SELECT *
    FROM db
    LIMIT 10
""", conn)
```

### 2c. Schema / Structure
> Shows all column names and their data types.

```python
pd.read_sql_query("""
PRAGMA table_info('db');
""", conn)
```

---
## Step 3 - Apply WHERE Filters

### 3a. Filter by Region = 'West'
> Retrieves all sales records from the West region.

```python
result = pd.read_sql_query("""
    SELECT *
    FROM db
    WHERE region = 'West'
""", conn)

print(f"Total rows (West region): {len(result)}")
result.head(10)
```

### 3b. Filter by Category = 'Furniture'
> Displays only records from the Furniture category.

```python
result = pd.read_sql_query("""
    SELECT *
    FROM db
    WHERE category = 'Furniture'
""", conn)

print(f"Total rows (Furniture): {len(result)}")
result.head(10)
```

### 3c. Filter by Sales > 500
> Identifies high-value transactions.

```python
result = pd.read_sql_query("""
    SELECT *
    FROM db
    WHERE sales > 500
""", conn)

print(f"Total rows (Sales > 500): {len(result)}")
result.head(10)
```

### 3d. Filter by Date - Year 2017
> Extracts all orders placed during 2017.

```python
result = pd.read_sql_query("""
SELECT *
FROM db
WHERE substr(order_date, -4) = '2017'
""", conn)

print(f"Total rows (Year 2017): {len(result)}")
result.head(10)
```

---
## Step 4 - GROUP BY Aggregations

### 4a. Total Sales by Region
> West leads with highest total sales.

```python
pd.read_sql_query("""
    SELECT region,
           ROUND(SUM(sales), 2) AS total_sales
    FROM db
    GROUP BY region
    ORDER BY total_sales DESC
""", conn)
```

### 4b. Total Quantity by Category
> Shows total units sold per category.

```python
pd.read_sql_query("""
    SELECT category,
           SUM(quantity) AS total_quantity
    FROM db
    GROUP BY category
    ORDER BY total_quantity DESC
""", conn)
```

### 4c. Average Sales by Category
> Technology has the highest average sales per transaction.

```python
pd.read_sql_query("""
    SELECT category,
           ROUND(AVG(sales), 2) AS avg_sales
    FROM db
    GROUP BY category
    ORDER BY avg_sales DESC
""", conn)
```

---
## Step 5 - Sort & Limit Results

### 5a. Top 10 Products by Sales
> Canon imageCLASS 2200 is the 1 revenue-generating product (~$61,599).

```python
pd.read_sql_query("""
    SELECT product_name,
           ROUND(SUM(sales), 2) AS total_sales
    FROM db
    GROUP BY product_name
    ORDER BY total_sales DESC
    LIMIT 10
""", conn)
```

### 5b. Top Categories by Profit
> Technology leads in profit, followed by Office Supplies. Furniture is the least profitable.

```python
pd.read_sql_query("""
    SELECT category,
           ROUND(SUM(profit), 2) AS total_profit
    FROM db
    GROUP BY category
    ORDER BY total_profit DESC
""", conn)
```

---
## Step 6 - Use Cases

### 6a. Monthly Profit Analysis
> Tracks how profit changes month by month across all years.

```python
pd.read_sql_query("""
    SELECT
        strftime('%Y-%m', order_date) AS month,
        ROUND(SUM(profit), 2)           AS total_profit
    FROM db
    GROUP BY month
    ORDER BY month
""", conn)
```

### 6b. Monthly Sales Analysis
> Evaluates monthly revenue trends over the full dataset period (2014–2017).

```python
pd.read_sql_query("""
    SELECT
        strftime('%Y-%m', order_date) AS month,
        ROUND(SUM(sales), 2)            AS total_sales
    FROM db
    GROUP BY month
    ORDER BY month
""", conn)
```

### 6c. Top 10 Customers by Sales


```python
pd.read_sql_query("""
    SELECT customer_name,
           COUNT(*)                  AS total_orders,
           ROUND(SUM(sales),  2)     AS total_sales,
           ROUND(SUM(profit), 2)     AS total_profit
    FROM db
    GROUP BY customer_name
    ORDER BY total_sales DESC
    LIMIT 10
""", conn)
```

### 6d. Duplicate Order Detection
> Checks for order IDs appearing more than once.

```python
result = pd.read_sql_query("""
    SELECT order_id,
           COUNT(*) AS count
    FROM db
    GROUP BY order_id
    HAVING COUNT(*) > 1
    ORDER BY count DESC
""", conn)

print(f"Order IDs with more than 1 row: {len(result)}")
result.head(10)
```

---
## Step 7 — Data Quality Checks

### 7a. Sales Equal to Zero
> Checks if any transaction has zero sales.

```python
result = pd.read_sql_query("""
    SELECT *
    FROM db
    WHERE sales = 0
""", conn)

if len(result) == 0:
    print("No records with zero sales.")
else:
    print(f"Found {len(result)} records with zero sales")
    result
```

### 7b. NULL Value Check
> Checks for missing values in critical columns.

```python
result = pd.read_sql_query("""
    SELECT *
    FROM db
    WHERE order_id IS NULL
       OR customer_name IS NULL
""", conn)

if len(result) == 0:
    print("No NULL values in order_id or customer_name.")
else:
    print(f"Found {len(result)} records with NULL values")
    result
```

### 7c. Final Row Count Validation
> Confirms total rows after all operations. 

```python
pd.read_sql_query("""
    SELECT COUNT(*) AS total_rows
    FROM db
""", conn)
```
